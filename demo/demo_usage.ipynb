```python
from collections import defaultdict

short_rules = set()
stats = defaultdict(int)

for rule in derivator.rules:
    if rule.name.split(':')[0] not in short_rules:
        stats[rule.pos_b, rule.pos_a, rule.info] += 1
        short_rules.add(rule.name.split(':')[0])
```
